```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range (1, 7):
    df = pd.read_csv("data-" + str(i) + ".csv")
    
    print("Data", i)
    
    print("X Variable")
    print("- Mean:", np.mean(df.x))
    print("- Standard Deviation:", np.std(df.x))
    print("- Range: [" + str(min(df.x)) + ", " + str(max(df.x)) + "]")
    
    print("Y Variable")
    print("- Mean:", np.mean(df.y))
    print("- Standard Deviation:", np.std(df.y))
    print("- Range: [" + str(min(df.y)) + ", " + str(max(df.y)) + "]")
    
    print("Correlation coefficient (r):", (df.x).corr(df.y))
    
    plt.plot(df.x, df.y, 'b.', alpha=0.5)
    plt.show()
    
    print("\n")
```
